1. Write a script to fetch data from a public API (e.g., GitHub API) and print it.

```python
import requests

def fetch_github_user(username):
    url = f"https://api.github.com/users/{username}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        print("Username:", data["login"])
        print("Name:", data["name"])
        print("Public repos:", data["public_repos"])
        print("Followers:", data["followers"])
        print("Profile URL:", data["html_url"])
    else:
        print("Error")


fetch_github_user("krishnadubey4397")
```

2. How do you pass parameters in a GET request using the requests library?

```python
import requests
para = {'hey': 20, 'hell': 77}
r = requests.get('https://httpbin.org/get', params=para)
print(r.json())
```
