```python
import time

def executiontime(fx):
    def mfx(*args, **kwargs):
        st = time.time()
        result = fx(*args, **kwargs)
        et = time.time()
        print(f"Execution time is : {et - st} sec")
        return result
    return mfx


@executiontime
def hello():
    print("hello everyone !")
    time.sleep(2)
    print("thank you !")

hello()
```
