```python

import time

def executiontime(func):
    def mfx():
        st = time.time()
        func()
        et = time.time()
        print(f"Execution time is : {et - st} sec")
    return mfx


@executiontime
def hello():
    print("hello everyone !")
    time.sleep(2)
    print("thank you !")

hello()
```
