P1 : Given two DataFrames, perform an outer join to combine them.

```python
import pandas as pd
import numpy as np


df1 = pd.DataFrame({
    "customer_id": [1, 2, 3, 4, 5, 6],
    "customer_name": ["Amit", "Neha", "Rahul", "Priya", "Sonal", "Karan"],
    "city": ["Indore", "Bhopal", "Delhi", "Mumbai", "Pune", "Jaipur"]
})

df2 = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105],
    "customer_id": [3, 4, 5, 7, 8],
    "order_amount": [250, 400, 150, 300, 500]
})

df3 = df1.merge(df2, how='outer', on='customer_id')
print(df3)
```

P2 : Create a pivot table to summarize sales data by region and product category.

```python
import pandas as pd

df = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105, 106, 107, 108],
    "region": ["North", "South", "North", "East", "West", "South", "East", "West"],
    "product_category": ["Electronics", "Furniture", "Furniture", "Electronics",
                         "Clothing", "Electronics", "Clothing", "Furniture"],
    "sales_amount": [50000, 30000, 20000, 45000, 15000, 40000, 18000, 22000]
})

df.pivot_table(index='region', columns='product_category', values='sales_amount')
```
